### Categorical Varaibles

Some of the columns are actually categorical variables but instead represented as numerical variables in the data frame. These columns are 'MSSubClass', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'MoSold', 'YrSold', 'GarageYrBlt'. We need to turn these columns into the `object` data type.

```python
ToBeCat = ['MSSubClass', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'MoSold', 'YrSold', 'GarageYrBlt']
data_train[ToBeCat] = data_train[ToBeCat].astype('str')

data_train.info()
```

### Numerical Variables

```python
data_train.describe()
```

### Missing Values

The missing values do have a meaning in the data set, for example, NA in the `Alley` column represents **no alley access**. Hence, we cannot just simply drop the columns with NA or drop rows with NA. First We need to convert the NAs in numerical variable columns to 0

```python
numericalNAcols = []
for col in cols:
    if data_train[col].dtype == 'float64' or data_train[col].dtype == 'int64':
        if data_train[col].isna().sum() > 0:
            numericalNAcols.append(col)
temp = data_train[numericalNAcols].fillna(0)
for col in numericalNAcols:
    data_train[col] = temp[col]
print(data_train.isna().sum())
```

For categorical variables, we can simply use the `get_dummies` function to take care of the NAs, as the `dummy_na` parameter is able to turn the NAs of each categorical variable into a category.

### Removal of Irelevant Columns

```python
data_train = data_train.drop(['Id'], axis = 1)
```

# Feature Engineering

### Dummy Encoding

```python
X = pd.get_dummies(data_train, dummy_na = True)

X
```

# Modelling
First, we are going to use gradient boosting as the model to predict house prices

```python
from sklearn.ensemble import GradientBoostingRegressor

Y = data['SalePrice']
model = GradientBoostingRegressor()
model.fit(X, Y)
model.score(X, Y)
```

This is a good fit on the training set, let's test it on the test set.

```python
data_test = pd.read_csv('housing_p_test.csv')
idtest = data_test['Id']
data_test[ToBeCat] = data[ToBeCat].astype('str')
data_test = data_test.drop(['Id'], axis = 1)
testcols = data_test.columns
numericalNAcolstest = []
for col in testcols:
    if data_test[col].dtype == 'float64' or data_test[col].dtype == 'int64':
        if data_test[col].isna().sum() > 0:
            numericalNAcolstest.append(col)
data_test[numericalNAcolstest] = data_test[numericalNAcolstest].fillna(0)

X_test = pd.get_dummies(data_test, dummy_na = True)

colsnotincluded = []
for col in X.columns:
    if col not in X_test.columns:
        colsnotincluded.append(col)

total_length_test = len(X_test)

new_X_test = X.iloc[:total_length_test, :]
new_X_test[colsnotincluded] = pd.DataFrame([[0] * len(colsnotincluded) for i in range(total_length_test)], columns=colsnotincluded)
duplicated_columns_to_remove_new = new_X_test[X_test.columns].columns.duplicated()
duplicated_columns_to_remove_old = X_test.columns.duplicated()
index_to_remove_new = []
index_to_remove_old = []
 
for i in range(len(duplicated_columns_to_remove_new)):
    if duplicated_columns_to_remove_new[i]:
        index_to_remove_new.append(new_X_test[X_test.columns].columns[i])
for i in range(len(duplicated_columns_to_remove_old)):
    if duplicated_columns_to_remove_old[i]:
        index_to_remove_old.append(X_test.columns[i])

new_X_test = new_X_test.drop(index_to_remove_new, axis = 1)
X_test = X_test.drop(index_to_remove_old, axis = 1)

new_X_test[X_test.columns] = X_test

new_X_train = X.drop(index_to_remove_new, axis = 1)

newmodel = GradientBoostingRegressor()
newmodel.fit(new_X_train, Y)
newmodel.score(new_X_train, Y)



predictions = newmodel.predict(new_X_test)

predictions_gb = pd.DataFrame({'Id': idtest, 'SalePrice': predictions})

predictions_gb.to_csv('gb.csv', index=False)
```

Next, I am going to use xgboost.

```python
from xgboost import XGBRegressor

best = {'colsample_bylevel': 0.6971640307744038, 'colsample_bytree': 0.6222251306452113, 'gamma': 0.045557016198868025, 'learning_rate': 0.012334455088591312, 'max_depth': 4.0, 'n_estimators': 2100.0, 'reg_lambda': 2.1713829044702413, 'subsample': 0.8459257473555255}
xgb_model = XGBRegressor(random_state=0,
                        n_estimators=int(best['n_estimators']), 
                        colsample_bytree= best['colsample_bytree'],
                        gamma= best['gamma'],
                        learning_rate= best['learning_rate'],
                        max_depth= int(best['max_depth']),
                        subsample= best['subsample'],
                        colsample_bylevel= best['colsample_bylevel'],
                        reg_lambda= best['reg_lambda'])
xgb_model.fit(new_X_train, Y)

xgb_model.score(new_X_train, Y)
```

This is even better than the gradient boost algorithm. Now we test it with the test data.

```python
predictions2 = xgb_model.predict(new_X_test)
predictions_xgb = pd.DataFrame({'Id': idtest, 'SalePrice': predictions2})

predictions_xgb.to_csv('xgb.csv', index=False)
```
